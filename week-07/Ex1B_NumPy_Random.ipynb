# Exercise 1.B NumPy Random

```python
import numpy as np
```

## Generating Random Numbers

```python
r_1 = np.random.randint(1, 20, size=2)
print(r_1)
```

```python
r_2 = np.random.randint(1, 20, size=(1, 2))
print(r_2)
```

```python
r_3 = np.random.random(size=(2, 1))
print(r_3)
```

```python
r_4 = np.random.randint(20, 100, size=(4, 5))
print(r_4)
```

```python
r_5 = np.random.random(size=(4, 5))
print(r_5)
```

## Generating Random Choices from an Array

```python
students = np.array(['Alice', 'Bob', 'Carlos', 'Diana', 'Eve', 'Frank',
                     'Grace', 'Hector', 'Iris', 'James', 'Karen', 'Leo'])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)
# Not ideal for breakout groups. The same student can appear more than once
```

```python
call_on_weighted = np.random.choice(students, size=5, p=[0.30, 0.10, 0.10, 0.10, 0.10,
                                                          0.05, 0.05, 0.05, 0.05, 0.05,
                                                          0.05, 0.00])
print(call_on_weighted)
# Alice has 30% chance, Leo has 0% chance, others share the remaining 70%
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
