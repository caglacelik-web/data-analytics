# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13], [14, 15, 16]])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

## The first loop prints each row multiplied by 2. The second loop prints each individual number multiplied by 2.

```python
for x in arrA:
    loop_result = x * 2
    print(loop_result)
# loop_result gets overwritten each loop. It only holds the last row's result
```

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
# This will throw an error — arrays must be the same shape to stack
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
# Error: all input arrays must have the same shape
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python
# This will error — axis=1 means columns must match in row count
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
# Error: arrays must have the same number of rows to join on axis=1
```

```python
h = np.hstack((arrA, arrA))   # side by side (like axis=1)
v = np.vstack((arrA, arrA))   # top to bottom (like axis=0)
d = np.dstack((arrA, arrA))   # depth to wise (adds a 3rd dimension)
print(h)
print(v)
print(d)
```

## Splitting Arrays

```python
splitsky = np.split(joined_array_1, 2)
print(splitsky)
# Result: 2 arrays, each with 2 rows (2D)
```

```python
spliterino = np.split(joined_array_1, 3)
print(spliterino)
# Error: 4 rows can't be divided equally into 3 parts
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
# arrA % 2 == 0 checks if a number is even (no remainder when divided by 2)
```

## Sorting and Filtering Arrays

```python
arrB = np.array([[7, 6], [13, 5], [7, 3]])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)
z = arrB < 7
arrC = arrB[z]
print(arrC)
```

```python
y = arrB >= 7
arrD = arrB[y]
print(arrD)
```
