# Exercise 2.A Fundamentals of pandas

```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))  #got an error so we had to convert set to list first
ser4 = pd.Series(ad)
print(ser1)
print(ser2)
print(ser3)
print(ser4)
```

```python
ser5 = pd.Series(17, 18, 19, 20)
# TypeError: got an unexpected keyword argument. Series doesn't accept multiple separate values like this
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 'col2', ser2)
df1.insert(2, 'col3', ser3)
df1
```

```python
df1['col4'] = ser4
df1
# ser4 has text index labels instead of numbers
# so pandas can't match them to df1's rows. The column fills with NaN
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```
