```python
import numpy as np
```

## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=2)
r_2 = np.random.randint(1, 20, size=(1, 2))
r_3 = np.random.random(size=(2, 1))
r_4 = np.random.randint(20, 100, size=(4, 5))
r_5 = np.random.random(size=(4, 5))
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
# shuffle mixes the ROWS but keeps each row intact
```

Shuffle rearranged the rows of r_4 but kept the values within each row in the same order.

```python
new_shuffle = np.random.permutation(r_4)
print(new_shuffle)
print(r_4)
# permutation returns a NEW shuffled array and r_4 itself stays unchanged
```

Use shuffle when you want to mix the original array in place.
Use permutation when you want a shuffled copy without changing the original.

## Mathematical Operations

```python
print(r_1, r_1.shape)
print(r_2, r_2.shape)
print(np.add(r_1, r_2))
print(np.add(r_1, r_2).shape)
```

```python
print(r_1, r_1.shape)
print(r_3, r_3.shape)
print(np.add(r_1, r_3))
print(np.add(r_1, r_3).shape)
```

```python
print(r_2, r_2.shape)
print(r_4, r_4.shape)
print(np.add(r_2, r_4))
print(np.add(r_2, r_4).shape)
# ValueError: r_2 has 2 columns and r_4 has 5 columns. Can't broadcast together
```

```python
# np.add(r_1, r_4) throws an error : shapes (2,) and (4,5) are not compatible
# np.add(r_2, r_3) also errors : (1,2) and (2,1) don't broadcast together
print(np.add(r_1, r_4))
```

```python
print(np.add(np.array([1]), r_4))
# Adds 1 to every element in r_4
```

```python
print(np.add(np.array([[1]]), r_4))
# Same result. [[1]] broadcasts across all rows and columns
```

```python
print(np.add(np.array([1,1,1,1,1]), r_4))
# Adds 1 to each column across every row
```

```python
print(np.add(np.array([[1],[1]]), r_4))
# Error: shape (2,1) can't broadcast with (4,5) — row count doesn't match
```

Broadcasting lets NumPy add arrays of different shapes by stretching the smaller
one to match the larger one. The shapes must be compatible. Dimensions need to
either match or one of them needs to be 1.

## Rounding

```python
print(r_3)
```

```python
print(np.trunc(r_3))   # drops the decimal, toward zero
```

```python
print(np.fix(r_3))     # same as trunc
```

```python
print(np.floor(r_3))   # rounds DOWN to nearest whole number
```

```python
print(np.ceil(r_3))    # rounds UP to nearest whole number
```

```python
print(np.round(r_3, 2))  # rounds to 2 decimal places
```

## Cumulative Operations

```python
print(r_4)
```

```python
print(np.sum(r_4))          # total of all values
```

```python
print(np.sum(r_4, axis=0))  # sum of each column
```

```python
print(np.sum(r_4, axis=1))  # sum of each row
```

```python
print(np.cumsum(r_4))       # rolling total of all values
```

```python
print(np.cumsum(r_4, axis=0))  # rolling total down each column
```

## Grocery Problem

```python
# Order: vegetables, fruits, snacks, beverages
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])
print(bag1)
print(bag2)
```

```python
totals = np.add(bag1, bag2)
print(f'The first bag contains {np.sum(bag1)} items, and the second bag contains {np.sum(bag2)} items.')
```

```python
print(f'There are {totals[0]} vegetables, {totals[1]} fruits, {totals[2]} boxes of snacks, and {totals[3]} beverages in the two grocery bags.')
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
```

```python
ben = np.array([0, 5, 5, 3, 4])
# Ben missed Monday so his value is 0
```

```python
carla = np.array([3, 6, 0, 3, 4])
# Carla was sick Wednesday so her value is 0
```

```python
all_knocks = np.array([ali, ben, carla])
print(all_knocks)
```

```python
print(f'Ali knocked on {np.sum(ali)} doors, Ben knocked on {np.sum(ben)}, Carla knocked on {np.sum(carla)}.')
print(f'Total doors knocked: {np.sum(all_knocks)}')
```

```python
daily_totals = np.sum(all_knocks, axis=0)
rolling = np.cumsum(daily_totals)
estimated_votes = rolling * 0.5
print(estimated_votes)
# Every 2 knocks = 1 estimated vote, so multiply rolling total by 0.5
```
