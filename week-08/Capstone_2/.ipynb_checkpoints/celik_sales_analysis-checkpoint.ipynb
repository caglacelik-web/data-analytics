```python
# Cleaning dates and data types
store_sales['Transaction Date'] = pd.to_datetime(store_sales['Transaction Date'])
store_sales['Year'] = store_sales['Transaction Date'].dt.year
store_sales['Month'] = store_sales['Transaction Date'].dt.month
store_sales['YearMonth'] = store_sales['Transaction Date'].dt.to_period('M')

# Fix customer list ID column name for merging later
customer_list = customer_list.rename(columns={'cust_id': 'RewardsID'})
customer_list['RewardsID'] = customer_list['RewardsID'].astype(float)

print("Date range:", store_sales['Transaction Date'].min(), "to", store_sales['Transaction Date'].max())
print("Total transactions:", len(store_sales))
```

**Territory Managers and Assigned Store Locations**

- This section identifies the territory managers, store IDs, and store locations included in each assigned sales territory.

```python
# Two territories
miami_vue_stores = store_detail[store_detail['Territory Manager'] == 'Miami Vue']
bo_heap_stores = store_detail[store_detail['Territory Manager'] == 'Bo Heap']

print("=" * 55)
print("MIAMI VUE — Northeast Territory")
print("=" * 55)
print(f"Region: {miami_vue_stores['Region'].iloc[0]}")
print(f"Region Director: {miami_vue_stores['Region Director'].iloc[0]}")
print(f"Number of Stores: {len(miami_vue_stores)}\n")
print(miami_vue_stores[['Store ID', 'Store Location', 'State']].to_string(index=False))

print("\n" + "=" * 55)
print("BO HEAP — Northeast Territory")
print("=" * 55)
print(f"Region: {bo_heap_stores['Region'].iloc[0]}")
print(f"Region Director: {bo_heap_stores['Region Director'].iloc[0]}")
print(f"Number of Stores: {len(bo_heap_stores)}\n")
print(bo_heap_stores[['Store ID', 'Store Location', 'State']].to_string(index=False))
```

### Territory Overview

Miami Vue manages 16 stores across New Jersey, while Bo Heap oversees 18 stores across Massachusetts.
Both territories operate within the Northeast region under Director Michael Jarvis,
making them a strong comparison for regional sales and marketing analysis. 
Because both territories serve densely populated metro areas, comparing their sales performance can help identify 
customer trends, high-performing stores, and opportunities for future marketing focus.

--------
**Monthly Sales Comparison by Territory**

- This analysis compares monthly in-store revenue between the Miami Vue and Bo Heap territories to identify sales trends and overall performance over time.

```python
# Filter sales to our two territories
miami_vue_ids = miami_vue_stores['Store ID'].tolist()
bo_heap_ids = bo_heap_stores['Store ID'].tolist()

sales_mv = store_sales[store_sales['Store ID'].isin(miami_vue_ids)].copy()
sales_bh = store_sales[store_sales['Store ID'].isin(bo_heap_ids)].copy()

print(f"Miami Vue transactions: {len(sales_mv):,}")
print(f"Bo Heap transactions: {len(sales_bh):,}")
```

```python
# Monthly revenue by territory
monthly_mv = sales_mv.groupby('YearMonth')['Sale Amount'].sum().reset_index()
monthly_bh = sales_bh.groupby('YearMonth')['Sale Amount'].sum().reset_index()

monthly_mv['YearMonth_dt'] = monthly_mv['YearMonth'].dt.to_timestamp()
monthly_bh['YearMonth_dt'] = monthly_bh['YearMonth'].dt.to_timestamp()

print("Miami Vue — Monthly Revenue (sample):")
print(monthly_mv.head(6).to_string(index=False))
```

```python
# CHART 1: Monthly Revenue - Miami Vue vs Bo Heap
fig, ax = plt.subplots(figsize=(13, 5))

ax.plot(monthly_mv['YearMonth_dt'], monthly_mv['Sale Amount'],
        color='#2563EB', linewidth=2.5, marker='o', markersize=3.5,
        label="Miami Vue (NJ)")

ax.plot(monthly_bh['YearMonth_dt'], monthly_bh['Sale Amount'],
        color='#DC2626', linewidth=2.5, marker='o', markersize=3.5,
        label="Bo Heap (Northeast)")

# Format y-axis as currency
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Format x-axis as readable dates
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45, ha='right')

ax.set_title("Monthly In-Store Revenue: Miami Vue vs Bo Heap\nNortheast Region, 2022–2025",
             pad=15)
ax.set_xlabel("Month")
ax.set_ylabel("Total Revenue")
ax.legend(loc='upper left', frameon=False)

# Drawing an arrow pointing to the peak month with adding an annotate. 
peak_idx = monthly_mv['Sale Amount'].idxmax()
peak_date = monthly_mv.loc[peak_idx, 'YearMonth_dt']
peak_val = monthly_mv.loc[peak_idx, 'Sale Amount']
ax.annotate(f"Peak: ${peak_val:,.0f}",
            xy=(peak_date, peak_val),
            xytext=(10, -30), textcoords='offset points',
            fontsize=9, color='#2563EB',
            arrowprops=dict(arrowstyle='->', color='#2563EB', lw=1))

plt.tight_layout()
plt.savefig('chart1_monthly_revenue.png', dpi=150, bbox_inches='tight')
plt.show()
```

-------
**Evaluating Store Performance Across Territories**
- This section evaluates store performance within each territory by comparing total sales revenue to identify top-performing locations and stores with potential opportunities for improvement.

```python
# Revenue by store - Miami Vue territory
store_revenue_mv = sales_mv.groupby('Store ID')['Sale Amount'].agg(
    Total_Revenue='sum',
    Transactions='count',
    Avg_Transaction='mean'
).reset_index()

# Join store names from store_detail
store_revenue_mv = store_revenue_mv.merge(
    store_detail[['Store ID', 'Store Location']], on='Store ID'
)

store_revenue_mv = store_revenue_mv.sort_values('Total_Revenue', ascending=False)
store_revenue_mv['Rank'] = range(1, len(store_revenue_mv) + 1)

print("Miami Vue Territory — Store Performance Rankings:")
print(store_revenue_mv[['Rank', 'Store Location', 'Total_Revenue',
                          'Transactions', 'Avg_Transaction']].to_string(index=False))
```

```python
# Revenue by store - Bo Heap territory
store_revenue_bh = sales_bh.groupby('Store ID')['Sale Amount'].agg(
    Total_Revenue='sum',
    Transactions='count',
    Avg_Transaction='mean'
).reset_index().merge(store_detail[['Store ID', 'Store Location']], on='Store ID')
store_revenue_bh = store_revenue_bh.sort_values('Total_Revenue', ascending=False)
store_revenue_bh['Rank'] = range(1, len(store_revenue_bh) + 1)

print("Bo Heap Territory — Store Performance Rankings:")
print(store_revenue_bh[['Rank', 'Store Location', 'Total_Revenue',
                          'Transactions', 'Avg_Transaction']].to_string(index=False))
```

```python
# CHART 2: Top 8 stores per territory (side by side)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Miami Vue — top 8
top_mv = store_revenue_mv.head(8)
bars1 = ax1.barh(top_mv['Store Location'][::-1],
                  top_mv['Total_Revenue'][::-1],
                  color='#2563EB', alpha=0.85)
ax1.set_title("Miami Vue\nTop Stores by Total Revenue", fontsize=13)
ax1.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))
ax1.set_xlabel("Total Revenue")

# Add value labels on bars
for bar in bars1:
    width = bar.get_width()
    ax1.text(width * 1.01, bar.get_y() + bar.get_height() / 2,
             f'${width:,.0f}', va='center', fontsize=9, color='#1e40af')

# Bo Heap — top 8
top_bh = store_revenue_bh.head(8)
bars2 = ax2.barh(top_bh['Store Location'][::-1],
                  top_bh['Total_Revenue'][::-1],
                  color='#DC2626', alpha=0.85)
ax2.set_title("Bo Heap\nTop Stores by Total Revenue", fontsize=13)
ax2.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))
ax2.set_xlabel("Total Revenue")

for bar in bars2:
    width = bar.get_width()
    ax2.text(width * 1.01, bar.get_y() + bar.get_height() / 2,
             f'${width:,.0f}', va='center', fontsize=9, color='#991b1b')

plt.suptitle("EmporiUm — Northeast Region Store Rankings", fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('chart2_store_rankings.png', dpi=150, bbox_inches='tight')
plt.show()
```

-----
**Top Rewards Customers by Sales Revenue**

- This analysis compares rewards member purchase activity across both territories to identify the highest-spending customers and evaluate customer contribution to overall sales revenue.

```python
# Top customers — Miami Vue territory
# Only transactions with a RewardsID (loyalty members)
cust_mv = sales_mv[sales_mv['RewardsID'].notna()].copy()

top_customers_mv = cust_mv.groupby('RewardsID')['Sale Amount'].agg(
    Total_Spend='sum',
    Transactions='count'
).reset_index().sort_values('Total_Spend', ascending=False)

# Join names from customer list
top_customers_mv = top_customers_mv.merge(
    customer_list[['RewardsID', 'name', 'email']], on='RewardsID', how='left'
)

print("Miami Vue Territory — Top 15 Rewards Customers:")
print(top_customers_mv.head(15)[['name', 'Total_Spend',
                                    'Transactions', 'email']].to_string(index=False))
```

```python
# Top customers — Bo Heap territory
# Only transactions with a RewardsID (loyalty members)

cust_bh = sales_bh[sales_bh['RewardsID'].notna()].copy()
top_customers_bh = cust_bh.groupby('RewardsID')['Sale Amount'].agg(
    Total_Spend='sum',
    Transactions='count'
).reset_index().sort_values('Total_Spend', ascending=False).merge(
    customer_list[['RewardsID', 'name', 'email']], on='RewardsID', how='left'
)

print("Bo Heap Territory — Top 15 Rewards Customers:")
print(top_customers_bh.head(15)[['name', 'Total_Spend',
                                    'Transactions', 'email']].to_string(index=False))
```

-----
**Revenue and Transaction Trends by Product Category**

- This analysis examines monthly transaction volume and total sales revenue by product category across both territories to identify the most popular product categories, the strongest revenue drivers, and potential opportunities for future marketing growth.

```python
# Merge sales → products → categories
sales_with_products = store_sales.merge(products, on='Prod Num', how='left')
sales_with_categories = sales_with_products.merge(
    product_categories, on=['CategoryID', 'SubcategoryID'], how='left'
)

# Filter to our territories
cat_mv = sales_with_categories[sales_with_categories['Store ID'].isin(miami_vue_ids)].copy()
cat_bh = sales_with_categories[sales_with_categories['Store ID'].isin(bo_heap_ids)].copy()
```

```python
# Monthly transactions by category — Miami Vue
cat_monthly_mv = cat_mv.groupby(['YearMonth', 'Category']).agg(
    Transactions=('Sale Amount', 'count'),
    Revenue=('Sale Amount', 'sum')
).reset_index()
cat_monthly_mv['YearMonth_dt'] = cat_monthly_mv['YearMonth'].dt.to_timestamp()

# Overall category totals — Miami Vue
cat_totals_mv = cat_mv.groupby('Category').agg(
    Transactions=('Sale Amount', 'count'),
    Revenue=('Sale Amount', 'sum')
).reset_index().sort_values('Revenue', ascending=False)

print("Miami Vue — Revenue and Transactions by Category:")
print(cat_totals_mv.to_string(index=False))
```

```python
# Monthly transactions by category — Bo Heap

cat_totals_bh = cat_bh.groupby('Category').agg(
    Transactions=('Sale Amount', 'count'),
    Revenue=('Sale Amount', 'sum')
).reset_index().sort_values('Revenue', ascending=False)

print("Bo Heap — Revenue and Transactions by Category:")
print(cat_totals_bh.to_string(index=False))
```

```python
# CHART 3: Category revenue — stacked comparison
categories = cat_totals_mv['Category'].tolist()
rev_mv = cat_totals_mv.set_index('Category')['Revenue']
rev_bh = cat_totals_bh.set_index('Category').reindex(categories)['Revenue'].fillna(0)

x = np.arange(len(categories))
width = 0.38

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, rev_mv, width, color='#2563EB', alpha=0.85,
               label='Miami Vue')
bars2 = ax.bar(x + width/2, rev_bh, width, color='#DC2626', alpha=0.85,
               label='Bo Heap')

ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=25, ha='right', fontsize=10)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))
ax.set_title("Revenue by Product Category\nMiami Vue vs Bo Heap — Full Period", pad=15)
ax.set_ylabel("Total Revenue")
ax.legend(frameon=False)

plt.tight_layout()
plt.savefig('chart3_category_revenue.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# CHART 4: Monthly transaction trend by top 3 categories — Miami Vue
top3_cats = cat_totals_mv.head(3)['Category'].tolist()
colors_cats = ['#2563EB', '#16A34A', '#D97706']

fig, ax = plt.subplots(figsize=(13, 5))

for cat, color in zip(top3_cats, colors_cats):
    df_cat = cat_monthly_mv[cat_monthly_mv['Category'] == cat]
    ax.plot(df_cat['YearMonth_dt'], df_cat['Transactions'],
            label=cat, color=color, linewidth=2, marker='o', markersize=3)

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45, ha='right')
ax.set_title("Monthly Transaction Volume — Top 3 Categories\nMiami Vue Territory", pad=15)
ax.set_ylabel("Number of Transactions")
ax.set_xlabel("Month")
ax.legend(frameon=False)

plt.tight_layout()
plt.savefig('chart4_category_trends.png', dpi=150, bbox_inches='tight')
plt.show()
```

### Marketing Recommendation — Next Quarter 2026 Focus
**Prepared by:** Cagla Celik | **For:** Miami Vue & Michael Jarvis, Northeast Region

---

#### Key Findings

From January 2022 to September 2025, Miami Vue’s territory generated over **$5.1M** in revenue across more than **38,000 transactions**. While Bo Heap’s territory performed slightly higher overall, the difference is largely explained by having more stores. Revenue growth in Miami Vue’s territory more than doubled between 2022 and 2025, showing strong upward momentum.

Newark and Trenton were the top-performing stores, together contributing nearly 24% of total territory revenue. In contrast, Paterson showed unusually low activity and stopped recording sales after November 2024, which may indicate an operational or reporting issue that should be reviewed.

Technology & Accessories was the strongest product category by far, generating nearly 70% of total territory revenue despite fewer transactions than lower-cost categories like Stationery. This suggests that high-ticket technology purchases are the primary driver of revenue growth.

Rewards members also spent more per transaction on average than non-members, highlighting the value of customer loyalty programs and repeat shoppers.

---

#### Recommendations for Next Quarter 2026

**1. Prioritize Technology & Accessories marketing.**  
This category drives the majority of revenue and presents the largest growth opportunity for Miami Vue’s territory. Focus promotions, product displays, and campaigns around tech products.

**2. Review the Paterson store immediately.**  
The sudden drop in activity after November 2024 should be investigated to determine whether it is a store closure, operational issue, or missing sales data.

**3. Expand rewards program enrollment.**  
Rewards members spend more per visit, making loyalty growth a cost-effective strategy for increasing repeat business and long-term customer value.

**4. Focus campaigns on top-performing stores first.**  
Newark and Trenton consistently lead the territory in both revenue and transaction volume, making them ideal locations for launching new marketing initiatives.
