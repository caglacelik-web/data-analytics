```python
import numpy as np
```

## Array Basics

```python
fruits = ['apple', 'banana', 'cherry', 'date']
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
print(arr_1.dtype)
# The letter 'U' means Unicode string (text data)
# The number after U is the length of the longest string in the array
```

```python
arr_2 = np.array([(10, 11, 12), (13, 14, 15)])
print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)
# No! .T does not change the original array, it just displays it transposed.
# arr_2 still has its original shape.
```

```python
arr_3 = np.array([(1, 3, 5), (2, 4, 6)])
print(arr_3)
```

```python
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
# arr_3[0,2] = 5 (row 0, column 2)
# arr_3[1,1] = 4 (row 1, column 1)
```

```python
arr_5 = np.ones((2, 2, 2))
print(arr_5)
# A 3D array with two values in each dimension, all filled with 1s
```
