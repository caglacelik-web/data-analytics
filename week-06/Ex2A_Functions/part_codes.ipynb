# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(part_code):
    return part_code[:part_code.index(':')]

def get_product_num(part_code):
    return part_code[part_code.index(':') + 1 : part_code.index('-')]

def get_size(part_code):
    return part_code[part_code.index('-') + 1:]
```

```python
code1 = 'ACME:123-L'
code2 = 'DI:12345-M'
code3 = 'ACE:1-12'
```

```python
for code in [code1, code2, code3]:
    supplier = get_supplier_code(code)
    product = get_product_num(code)
    size = get_size(code)
    print(f"For {code}, the supplier is {supplier}. Part number {product} is available in size {size}.")
```
