# Exercise 2.C Creating and calling functions
## Lab 1. Working with defining custom functions

```python
def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician is {music}."

print(favorite_things('Cagla', 'Hereditary', 'Doja Cat'))
```

```python
def why_im_here(purpose):
    return purpose

print(why_im_here('I joined Year Up United to build my data analytics skills.'))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."

print(favorite_place('Istanbul', 'it has amazing food and rich history'))
```

```python
print(favorite_place('Chicago', 'it has beautiful architecture and great energy'))
```
