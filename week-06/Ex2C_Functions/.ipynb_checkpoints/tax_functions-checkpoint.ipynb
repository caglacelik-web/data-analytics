# Exercise 2.C Creating and calling functions
## Lab 3. Tax functions
This notebook demonstrates custom functions for calculating payroll tax withholdings
including Social Security, Medicare, and Federal income tax.

```python
def get_ssn_tax(gross_pay):
    return gross_pay * 0.062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * 0.0145
```

```python
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        rate = 0.23
    elif withholding_code == 1:
        rate = 0.21
    elif withholding_code == 2:
        rate = 0.195
    elif withholding_code == 3:
        rate = 0.185
    else:
        rate = 0.18 - ((withholding_code - 4) * 0.005)
    return gross_pay * rate
```

```python
gross1 = 750
code1 = 0
print(f"Person 1 | Gross Pay: ${gross1}")
print(f"Social Security Tax: ${get_ssn_tax(gross1):.2f}")
print(f"Medicare Tax: ${get_medicare_tax(gross1):.2f}")
print(f"Federal Tax: ${get_federal_tax(gross1, code1):.2f}")
```

```python
gross2 = 1550
code2 = 2
print(f"Person 2 | Gross Pay: ${gross2}")
print(f"Social Security Tax: ${get_ssn_tax(gross2):.2f}")
print(f"Medicare Tax: ${get_medicare_tax(gross2):.2f}")
print(f"Federal Tax: ${get_federal_tax(gross2, code2):.2f}")
```
