## Pi Birthday Check
This notebook checks whether your birthday appears in the first million digits of pi.

```python
pi = open('pi_million_digits.txt', 'r')
print(pi.readline())
pi.close()
```

```python
pi = open('pi_million_digits.txt', 'r')
pi_lines = pi.readlines()
pi.close()
```

```python
def pi_bday_check(bday):
    for i in pi_lines:
        if bday in i:
            print("Your birthday is in the first million digits of pi")
            break

pi_bday_check('071902')
```

```python
def pi_bday_check2(bday):
    birthday_found = 0
    for i in pi_lines:
        if bday in i:
            print("Your birthday is in the first million digits of pi")
            birthday_found += 1
            break
    if birthday_found != 1:
        print("Sorry, your birthday was not found in the first million digits of pi")

pi_bday_check2('071902')  # not found
pi_bday_check2('070281')  # found example
```

```python
pi_string = ''
for i in pi_lines:
    pi_string += i.strip()

print(pi_string[0:300])
```

```python
birthday_position = pi_string.index('070281') - 1
# We subtract 1 because indexing starts at 0, and pi_string starts with '3.'
# so the actual decimal position is one less than the index
print(birthday_position)
```
